Q6. Convert all the files of Q4 into upper case parallel using multi-threading.

```python
import concurrent.futures

# Target file sizes in bytes (100MB, 200MB, 300MB, 400MB, 500MB)
target_file_sizes = [100 * 1024 * 1024, 200 * 1024 * 1024, 300 * 1024 * 1024, 400 * 1024 * 1024, 500 * 1024 * 1024]

# Length of each string
string_length = 100

# Function to convert file contents to uppercase
def convert_to_uppercase(file_name):
    with open(file_name, 'r+') as file:
        content = file.read().upper()
        file.seek(0)
        file.write(content)
        file.truncate()
    return file_name

# Generate and write random strings to multiple files
with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = []
    for index, target_file_size in enumerate(target_file_sizes, start=1):
        # Calculate the number of lines required to achieve the target file size
        num_lines = target_file_size // (string_length + 1)  # +1 for the newline character

        random_strings = [generate_random_string(string_length) for _ in range(num_lines)]
        file_name = f"random_strings_{index}GB.txt"

        with open(file_name, "w") as file:
            file.write('\n'.join(random_strings))

        futures.append(executor.submit(convert_to_uppercase, file_name))

    for future in concurrent.futures.as_completed(futures):
        result = future.result()
        print(f"File '{result}' converted to uppercase successfully.")
```

```python
import requests
from googlesearch import search

def download_images(keyword, limit):
    image_urls = []
    for url in search(keyword + " cat", num_results=limit):
        if url.endswith(('.jpg', '.jpeg', '.png', '.gif')):
            image_urls.append(url)

    count = 0
    for url in image_urls:
        try:
            response = requests.get(url)
            response.raise_for_status()
            with open(f"cat_images/cat{count+1}.jpg", "wb") as f:
                f.write(response.content)
            print(f"Image {count+1} downloaded successfully.")
            count += 1
            if count == limit:
                break
        except requests.exceptions.HTTPError as e:
            print(f"Error downloading image from URL: {url}")
            continue

download_images("cat", 10)
```

Q10. Create an automated pipeline using multi-threading for:
“Automatic Download of 100 Videos from YouTube” → “Convert it to Audio”.

```python
pip install pytube
pip install moviepy
import os
import threading
from pytube import YouTube
from moviepy.editor import VideoFileClip

# Create a directory to save the downloaded videos and converted audio files
output_dir = "youtube_pipeline"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Define the search query and the number of videos to download
search_query = "Machine Learning"
num_videos = 100

# Create a lock to synchronize access to the output directory
lock = threading.Lock()

# Function to download a video from YouTube
def download_video(url, output_path):
    try:
        # Create a YouTube object
        yt = YouTube(url)

        # Select the highest resolution available for download
        video_stream = yt.streams.get_highest_resolution()

        # Download the video
        video_stream.download(output_path)

        # Print the video title and the downloaded file path
        print(f"Downloaded video: {yt.title}")
        print(f"Saved as: {output_path}")

        # Convert the downloaded video to audio
        convert_to_audio(output_path)
    except Exception as e:
        print(f"Error occurred while downloading {url}: {str(e)}")

# Function to convert a video to audio
def convert_to_audio(video_path):
    try:
        # Load the video clip
        video_clip = VideoFileClip(video_path)

        # Generate the output audio file path
        audio_filename = os.path.splitext(os.path.basename(video_path))[0] + ".mp3"
        audio_path = os.path.join(output_dir, audio_filename)

        # Convert the video clip to audio
        video_clip.audio.write_audiofile(audio_path)

        # Print the audio file path
        print(f"Converted to audio: {audio_path}")

        # Delete the video file
        os.remove(video_path)
    except Exception as e:
        print(f"Error occurred while converting {video_path} to audio: {str(e)}")

# Perform the search on YouTube
search_results = YouTube.search(search_query, num_results=num_videos)

# Download and convert videos using multiple threads
for video in search_results:
    # Get the YouTube video URL
    video_url = f"https://www.youtube.com/watch?v={video.video_id}"

    # Create a thread for downloading and converting the video
    thread = threading.Thread(target=download_video, args=(video_url, output_dir))
    thread.start()

    # Wait for the thread to complete before starting the next one
    thread.join()

print("Pipeline completed!")
```

Q11. Create an automated pipeline using multi-threading for: “Automatic Download of 500 images of Dog from
GoogleImages” → “Rescale it to 50%”.

```python
10.
import os
import requests
import threading
import concurrent.futures
from PIL import Image

# Function to download an image from a given URL
def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)

# Function to rescale an image to 50%
def rescale_image(image_path, save_path):
    image = Image.open(image_path)
    width, height = image.size
    new_width = int(width * 0.5)
    new_height = int(height * 0.5)
    resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)
    resized_image.save(save_path)

# Main function for the automated pipeline
def automated_pipeline():
    # Create a directory to save the downloaded images
    save_directory = "dog_images"
    if not os.path.exists(save_directory):
        os.makedirs(save_directory)

    # Fetch URLs of dog images from Google Images
    # You can modify the search query to customize the image results
    search_query = "dog"
    url = f"https://www.google.com/search?q={search_query}&tbm=isch"

    response = requests.get(url)
    image_urls = response.text.split('src="')[1:]

    image_count = 0

    # Create a thread pool for concurrent execution
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Download and rescale each image using multiple threads
        for image_url in image_urls:
            if image_count >= 500:
                break

            image_url = image_url.split('"')[0]
            image_extension = image_url.split(".")[-1]

            if image_extension.lower() not in ["jpg", "jpeg", "png"]:
                continue

            save_path = os.path.join(save_directory, f"image_{image_count}.{image_extension}")

            executor.submit(download_image, image_url, save_path)
            executor.submit(rescale_image, save_path, save_path)

            image_count += 1

    print(f"Downloaded and rescaled {500} images.")

# Run the automated pipeline
automated_pipeline()
```

Q13. Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, pairwise_distances

# Set random seed for reproducibility
np.random.seed(123)

# Create a random dataset of 500 rows and 10 columns
dataset = pd.DataFrame({
    'col1': np.random.uniform(-10, 10, size=500),
    'col2': np.random.uniform(-10, 10, size=500),
    'col3': np.random.uniform(-10, 10, size=500),
    'col4': np.random.uniform(-10, 10, size=500),
    'col5': np.random.uniform(10, 20, size=500),
    'col6': np.random.uniform(10, 20, size=500),
    'col7': np.random.uniform(10, 20, size=500),
    'col8': np.random.uniform(10, 20, size=500),
    'col9': np.random.uniform(-100, 100, size=500),
    'col10': np.random.uniform(-100, 100, size=500)
})

# Apply K-Means clustering
kmeans_clusters = []
kmeans_scores = []
max_clusters = 10  # maximum number of clusters to try

for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(dataset)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(dataset, labels)
    kmeans_clusters.append(n_clusters)
    kmeans_scores.append(silhouette_avg)

# Plot K-Means clustering distance metric graph
plt.plot(kmeans_clusters, kmeans_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('K-Means Clustering - Silhouette Score vs Number of Clusters')
plt.show()

# Apply Hierarchical clustering
hierarchical_clusters = []
hierarchical_scores = []

for n_clusters in range(2, max_clusters + 1):
    hierarchical = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
    hierarchical.fit(dataset)
    labels = hierarchical.labels_
    silhouette_avg = silhouette_score(dataset, labels)
    hierarchical_clusters.append(n_clusters)
    hierarchical_scores.append(silhouette_avg)

# Plot Hierarchical clustering distance metric graph
plt.plot(hierarchical_clusters, hierarchical_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Hierarchical Clustering - Silhouette Score vs Number of Clusters')
plt.show()
```

Q15. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
import numpy as np
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Set the random seed for reproducibility
np.random.seed(42)

# Generate the random dataset
dataset = np.random.uniform(5, 10, size=(500, 5))

# Perform t-Tests on each column
for i in range(dataset.shape[1]):
    col_data = dataset[:, i]
    t_statistic, p_value = ttest_1samp(col_data, 5.0)
    print(f"Column {i+1} - t-Test:")
    print(f"   t-statistic: {t_statistic:.4f}")
    print(f"   p-value: {p_value:.4f}")

# Perform Wilcoxon Signed Rank Tests on each column
for i in range(dataset.shape[1]):
    col_data = dataset[:, i]
    statistic, p_value = wilcoxon(col_data - 5.0)
    print(f"Column {i+1} - Wilcoxon Signed Rank Test:")
    print(f"   statistic: {statistic:.4f}")
    print(f"   p-value: {p_value:.4f}")

# Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
col_3 = dataset[:, 2]
col_4 = dataset[:, 3]

t_statistic, p_value_ttest = ttest_ind(col_3, col_4)
statistic, p_value_wilcoxon = ranksums(col_3, col_4)

print("Two Sample t-Test:")
print(f"   t-statistic: {t_statistic:.4f}")
print(f"   p-value: {p_value_ttest:.4f}")

print("Wilcoxon Rank Sum Test:")
print(f"   statistic: {statistic:.4f}")
print(f"   p-value: {p_value_wilcoxon:.4f}")
```
